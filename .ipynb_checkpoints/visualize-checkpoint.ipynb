## training curve

```python
import pickle
with open('./curvelist.pkl','rb') as f:
    curves = pickle.load(f)
```

```python
def smooth(curve, w_len):
    c_len = len(curve)
    smoothed = []
    for i in range(c_len-w_len):
        smoothed.append(sum(curve[show_channel][i:i+w_len])/w_len)
    return smoothed
```

```python
import matplotlib.pyplot as plt
show_channel = 3
time_line = list(range(c_len))
smoothed = smooth(curves[show_channel])
delta = []
w_len = 100
for i in range(c_len):
    delta.append(curves[3][i]-curves[1][i])
plt.plot(time_line[:c_len-w_len], smoothed)
plt.figure()
plt.plot(time_line,delta)
```
