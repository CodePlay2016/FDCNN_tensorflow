## training curve

```python
import pickle
with open('./curvelist.pkl','rb') as f:
    curves = pickle.load(f)
```

```python
def smooth(curve, w_len):
    c_len = len(curve)
    smoothed = []
    for i in range(c_len-w_len):
        smoothed.append(sum(curve[i:i+w_len])/w_len)
    return smoothed
```

```python
import matplotlib.pyplot as plt
show_channel = 3
w_len = 1
time_line = list(range(c_len))

smoothed = smooth(curves[show_channel], w_len)
delta = []
for i in range(c_len):
    delta.append(abs(curves[3][i]-curves[1][i]))
plt.plot(time_line[:c_len-w_len], smoothed)
plt.figure()
plt.plot(time_line,curves[1])
```
